##BackPropagation

###Importing Libraries

```python
import numpy as np #array Operations
import pandas as pd #Handling Data
from sklearn.datasets import load_iris #Plant Iris Dataset
from sklearn.model_selection import train_test_split #Splitting Dataset into Train & Test
import matplotlib.pyplot as plt #plotting Graph
```

###Load Dataset

```python
dataset = load_iris()
dataset
```

###Segregating Data into X(Imput) & Y(Output)

```python
x=dataset.data
y=dataset.target
print(x)
print(y)
```

###Convert categorical variable into dummy/indicator variables

```python
y = pd.get_dummies(y).values
y 
```

###Split dataset into Train & Test

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=20, random_state=4)
```

###Initializing Hyper Parameters

```python
learningRate = 0.1
epoch = 5000
N = y_train.shape[0]
input_size = 4
hidden_size = 2 
output_size = 3  
results = pd.DataFrame(columns=["MeanSquareError", "accuracy"])
```

###Initializing Weights

###Activation Function

```python
def activationFn(x):
    return 1 / (1 + np.exp(-x))
```

###Mean Squared Error

```python
def mse(y_pred, y_true):
    return ((y_pred - y_true)**2).sum() / (2*y_pred.size)
```

###Evaluation Function - Accuracy

```python
def accuracy(y_pred, y_true):
    acc = y_pred.argmax(axis=1) == y_true.argmax(axis=1)
    return acc.mean()
```

```python
np.random.seed(10)
firstWeights = np.random.normal(scale=0.5, size=(input_size, hidden_size))   
lastWeights = np.random.normal(scale=0.5, size=(hidden_size , output_size)) 

print(firstWeights)
print(lastWeights)
```

###Training

```python
for i in range(epoch):
  firstNetInput= np.dot(X_train, firstWeights)
  firstAFnop=activationFn(firstNetInput)

  lastNetInput= np.dot(firstAFnop, lastWeights)
  lastAFnop=activationFn(lastNetInput)

# Calculating error
  mseValue = mse(lastAFnop, y_train)
  acc = accuracy(lastAFnop, y_train)
  results=results.append({"MeanSquareError":mseValue, "accuracy":acc},ignore_index=True)
    
# backpropagation
  E1 = lastAFnop - y_train
  dlastWeight = E1 * lastAFnop * (1 - lastAFnop)

  E2 = np.dot(dlastWeight, lastWeights.T)
  dfirstWeight = E2 * firstAFnop * (1 - firstAFnop)

  # weight updates
  lastWeight_update = np.dot(firstAFnop.T, dlastWeight) / N
  firstWeight_update = np.dot(X_train.T,  dfirstWeight) / N

  lastWeights = lastWeights - learningRate * lastWeight_update
  firstWeights = firstWeights - learningRate * firstWeight_update
```

```python
results.MeanSquareError.plot(title="Mean Squared Error")
```

```python
results.accuracy.plot(title="Accuracy")
```

```python
# feedforward
firstNetInput = np.dot(X_test, firstWeights)
firstAFnOp = activationFn(firstNetInput)

lastNetInput = np.dot(firstAFnOp, lastWeights)
lastAFnOp = activationFn(lastNetInput)

acc = accuracy(lastAFnOp, y_test)
print("Accuracy: {}".format(acc*100))
```
